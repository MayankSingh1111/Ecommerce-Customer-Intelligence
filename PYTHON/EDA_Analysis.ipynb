# 📊 E-Commerce Customer Intelligence  
## Exploratory Data Analysis (EDA)

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

df = pd.read_csv(r"E:\PROJECT\Ecommerce-Customer-Intelligence\DATA\PROCESSED\cleaned_data.csv")
chart_path = r"E:\PROJECT\Ecommerce-Customer-Intelligence\OUTPUT\CHARTS"
os.makedirs(chart_path, exist_ok=True)

df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df['date'] = pd.to_datetime(df['date'])

# save charts function
def save_plot(name):
    file_path = os.path.join(chart_path, f"{name}.png")
    plt.savefig(file_path, bbox_inches='tight')
    print(f"Saved: {file_path}")
```

# REVENUE TREND

```python
df['year_month'] = df['date'].dt.to_period('M')

monthly_revenue = df.groupby('year_month')['total_price'].sum()

plt.figure()
monthly_revenue.plot()
plt.title("Monthly Revenue Trend")
plt.xlabel("Time")
plt.ylabel("Revenue")
save_plot("Revenue_Trend")
plt.show()
```

# TOP PRODUCTS

```python
top_products = df.groupby('item_name')['total_price'].sum().sort_values(ascending=False).head(10)

plt.figure()
top_products.plot(kind='bar')
plt.title("Top 10 Products by Revenue")
save_plot("Top_Products")
plt.show()
```

# Top Customers

```python
top_customers = df.groupby('name')['total_price'].sum().sort_values(ascending=False).head(10)

plt.figure()
top_customers.plot(kind='bar')
plt.title("Top Customers")
save_plot("Top Customers")
plt.show()
```

# PAYMENT METHOD ANALYSIS

```python
payment = df['trans_type'].value_counts()

plt.figure()
payment.plot(kind='pie', autopct='%1.1f%%')
plt.title("Payment Method Distribution")
plt.show()
save_plot("Payment Method Distribution")
```

# SALES BY REGION

```python
region = df.groupby('division')['total_price'].sum().sort_values(ascending=False)

plt.figure()
region.plot(kind='bar')
plt.title("Revenue by Division")
plt.show()
save_plot("Sale By Region")
```

# PEAK SALES HOUR

```python
hour_sales = df.groupby('hour')['total_price'].sum()

plt.figure()
hour_sales.plot()
plt.title("Sales by Hour")
plt.show()
save_plot("Peak Sales Hour")
```

# CUSTOMER SEGMENTATION

```python
customer_spend = df.groupby('customer_key')['total_price'].sum()

segments = pd.cut(customer_spend,
                  bins=[0, 5000, 10000, float('inf')],
                  labels=['Low', 'Regular', 'VIP'])

segments.value_counts().plot(kind='bar')
plt.title("Customer Segmentation")
plt.show()
save_plot("Customer Segmentation")
```

## 📊 Key Insights

* **Consistent Revenue Growth:**
  Monthly revenue analysis shows a clear upward trend, indicating strong business expansion and increasing customer demand over time.

* **Revenue Concentration (Pareto Effect):**
  A small group of products contributes a large portion of total revenue, highlighting an 80/20 distribution where top products dominate sales.

* **High Customer Dependency:**
  A limited number of high-value (VIP) customers generate a significant share of total revenue, suggesting dependency on a small customer segment.

* **Dominance of Digital Payments:**
  Card-based transactions are the most preferred payment method, reflecting customer inclination toward digital payment systems.

* **Time-Based Sales Pattern:**
  Sales peak during specific hours of the day, indicating strong time-driven purchasing behavior that can be leveraged for targeted promotions.

* **Regional Performance Differences:**
  Certain divisions outperform others in terms of revenue, highlighting opportunities for targeted regional growth strategies.

---

## 📈 Conclusion

The business demonstrates strong growth potential but is heavily dependent on top-performing products and high-value customers. This indicates both strength and risk, as any fluctuation in these areas can significantly impact overall performance.
